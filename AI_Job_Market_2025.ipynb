# **Handling Missing Values**

```python
missing_report = df.isnull().sum()
print("\nMissing Values Summary:\n", missing_report[missing_report > 0])

df.dropna(subset=['job_id'], inplace=True)

for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype in ['float64', 'int64']:
            median_value = df[col].median()
            df[col].fillna(median_value, inplace=True)
            print(f"Filled missing numeric values in '{col}' with median: {median_value}")
        else:
            mode_value = df[col].mode()[0]
            df[col].fillna(mode_value, inplace=True)
            print(f"Filled missing categorical values in '{col}' with mode: {mode_value}")
```

# **Handling Duplicates**

```python
duplicates_jobid = df.duplicated(subset=['job_id']).sum()
print(f"\nDuplicates based on 'job_id': {duplicates_jobid}")

df.drop_duplicates(subset=['job_id'], inplace=True)

duplicates_other = df.duplicated(subset=['job_title', 'company_location', 'posting_date']).sum()
print(f"Additional duplicates based on job_title, company_location, posting_date: {duplicates_other}")
df.drop_duplicates(subset=['job_title', 'company_location', 'posting_date'], inplace=True)

print(f"Dataset shape after cleaning duplicates: {df.shape}")
```

# **Outlier Detection and Handling**

```python
# Calculate IQR for salary_usd
Q1 = df['salary_usd'].quantile(0.25)
Q3 = df['salary_usd'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"\nSalary IQR-based outlier bounds: lower={lower_bound}, upper={upper_bound}")

# Flag outliers
outliers = df[(df['salary_usd'] < lower_bound) | (df['salary_usd'] > upper_bound)]
print(f"Number of outlier salaries: {outliers.shape[0]}")

# Optionally remove outliers or cap values - here we cap to bounds
df['salary_usd_capped'] = np.where(df['salary_usd'] < lower_bound, lower_bound,
                          np.where(df['salary_usd'] > upper_bound, upper_bound, df['salary_usd']))
```

# **Top 10 Job Titles in AI Market**

```python
top_jobs = df['job_title'].value_counts().nlargest(10)
plt.figure(figsize=(12,6))
plt.plot(top_jobs.index, top_jobs.values, marker='o', linestyle='-', color='blue')
plt.title("Top 10 Job Titles in AI Market")
plt.xlabel("Job Title")
plt.ylabel("Number of Listings")
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# **Common Job Requirements**

```python
all_reqs = df["required_skills"].str.split(",").explode()
all_reqs = all_reqs.str.strip()
req_counts = all_reqs.value_counts().reset_index()
req_counts.columns = ['Skill', 'Count']

plt.figure(figsize=(12,6))
sns.barplot(data=req_counts, x='Skill', y='Count', palette='Blues_r')
plt.title(" Common Job Requirements", fontsize=16)
plt.xlabel("Requirement")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# **Heatmap Job Postings**

```python
pivot_table = pd.pivot_table(df, values='job_id', index='company_location', columns='remote_ratio', aggfunc='count', fill_value=0)
pivot_sample = pivot_table.sort_values(by=0, ascending=False).head(15)  # Top 15 countries

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_sample, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Jobdr Postings by Country and Remote Ratio')
plt.xlabel('Remote Ratio')
plt.ylabel('Company Location')
plt.tight_layout()
plt.show()
```
